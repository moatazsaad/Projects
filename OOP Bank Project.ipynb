```python
# Creating class with attrs name,age,gender,balance with a default value = 0
# Creating method to print the user's data
class User:
    
    def __init__(self,name,age,gender,balance=0):
        self.name = name
        self.age = age
        self.gender = gender
        self.balance = balance
    
    def show_data (self):
        print(f'Name: {self.name}')
        print(f'Age: {self.age}')
        print(f'Gender: {self.gender}')
        print(f'Balance: {self.balance}')
```

```python
user1 = User('Ahmed','20','Male',1000)
```

```python
user1.show_data()
```

```python
user2 = User('Sara','25','Female')
```

```python
user2.show_data()
```

```python
#Creating child class which inherit from the user class with methods(Deposite,Withdraw,Balance)
class Bank(User):
    
    def __init__(self, name, age, gender, balance, bank_name):
        if balance >= 0:
            super(Bank, self).__init__(name, age, gender, balance) 
            self.bank_name = bank_name
        else:
            raise Exception("Number should be positive")
    
    def show_data(self):
        super(Bank, self).show_data()
        print(f"Bank Name: {self.bank_name}")
        
    def deposite(self, amount):
        if amount > 0:
            self.balance = self.balance + amount
        else:
            print("Enter a valid positive amount")
    
    def show_balance(self):
        return self.balance
    
    def withdraw(self, amount):
        if amount > self.balance:
            print("Not enough money")
        elif amount <= 0:
                print("Insufficient withdraw amount, please enter a number greater than zero")
        else:
            self.balance = self.balance - amount
            print (f"Current balance:{self.show_balance()}")
```
